```python
import os

def zero_pad_png_in_nested_dirs(root_dir):
    """
    Zero-pad .png filenames in integer-named subdirectories within a root directory.
    
    Parameters:
        root_dir (str): Root directory containing the nested folder structure.
    """
    for dirpath, dirnames, filenames in os.walk(root_dir):
        # Process only integer-named subdirectories
        if os.path.basename(dirpath).isdigit():
            png_filenames = [f for f in filenames if f.endswith(".png") and f[:-4].isdigit()]
            
            if png_filenames:
                # Determine the maximum number of digits needed
                max_digits = len(str(max(map(int, [f[:-4] for f in png_filenames]))))
                
                # Rename files with zero-padding
                for filename in png_filenames:
                    name_part = filename[:-4]  # Remove .png extension
                    old_path = os.path.join(dirpath, filename)
                    new_filename = f"{int(name_part):0{max_digits}d}.png"
                    new_path = os.path.join(dirpath, new_filename)

                    # Rename the file
                    os.rename(old_path, new_path)
                    print(f"Renamed: {old_path} -> {new_path}")

# Example usage
root_directory = "res"
zero_pad_png_in_nested_dirs(root_directory)
```
